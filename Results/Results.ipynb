```python
#Load all data
data_s0 = pd.read_json('JSON/Avg/Wins Avg Single Q-Learning Type 0.json')
data_s1 = pd.read_json('JSON/Avg/Wins Avg Single Q-Learning Type 1.json')
data_d0 = pd.read_json('JSON/Avg/Wins Avg Double Q-Learning Type 0.json')
data_d1 = pd.read_json('JSON/Avg/Wins Avg Double Q-Learning Type 1.json')

#Get results
wins_s0 = data_s0['results'].values.tolist()
wins_s1 = data_s1['results'].values.tolist()
wins_d0 = data_d0['results'].values.tolist()
wins_d1 = data_d1['results'].values.tolist()

wins = [wins_s0, wins_s1, wins_d0, wins_d1]

for i, win in enumerate(wins):
    split = []

    start = 0
    end = 10

    while end <= 50:
        split.append(win[start:end])
        start += 10
        end += 10

    for j, arr in enumerate(split):
        split[j] = round(np.nanmean(arr) / 3, 0)

    wins[i] = split

train = ["20%", "40%", "60%", "80%", "100%"]

models = {
    "Single Type 0": wins[0],
    "Single Type 1": wins[1],
    "Double Type 0": wins[2],
    "Double Type 1": wins[3]
}

x = np.arange(len(train))

width = 0.2
multiplier = 0

fig, plt = plt.subplots(figsize=(15, 10))

for model, wins in models.items():
    offset = width * multiplier
    rects = plt.bar(x + offset, wins, width, label=model)
    multiplier += 1

plt.set_ylabel('Win %')
plt.set_xlabel('Percentage of training completed')
plt.set_title('Wins per percentage of training completed')
plt.set_xticks(x + width, train)
plt.legend(loc='upper left', ncols=4, prop={'size': 18})
plt.set_ylim(0, 100)

plt.show()
```
