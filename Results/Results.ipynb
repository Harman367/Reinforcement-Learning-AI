# Results Handling

```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load results
data = pd.read_json('VS Max Damage Player.json')
results_max = data['results']

#Plot results
plt.figure(figsize=(10, 5))
plt.xlabel('Number of battles')
plt.ylabel('Accumalted Reward')
plt.plot(results_max)
plt.legend()
plt.show()
```

```python
#Load results
data = pd.read_json('VS Max Damage Player.json')
results_max = data['results']

data = pd.read_json('VS Random Player.json')
results_random = data['results']

data = pd.read_json('Self Play.json')
results_self = data['results']

x = [*range(1, 11)]

#Plot results
plt.figure(figsize=(10, 5))
plt.xlabel('Every 100 out of 1000 games')
plt.ylabel('Number of wins')
plt.plot(x, results_max, label='Max Damage Player')
plt.plot(x, results_random, label='Random Player')
plt.plot(x, results_self, label='Self')
plt.legend()
plt.show()
```

# Win Rate
Max: AI player won 608 / 1000 battles [this took 93.41 seconds] <br>
Random: AI player won 922 / 1000 battles [this took 87.9 seconds] <br>
Self: AI player won 517 / 1000 battles [this took 109.52 seconds] <br>

```python
#Load results
data = pd.read_json('Accumulated Reward.json')
results = data['results']

title = "Learning Curve"

#Plot results
plt.figure(figsize=(10, 5))
plt.xlabel('Number of games')
plt.ylabel('Accumulated reward')
plt.plot(results, label=title)
plt.legend()
plt.show()
```
