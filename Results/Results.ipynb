# Results Handling

```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load results
data = pd.read_json('Accumulated Reward.json')
results = data['results']

#Plot results
plt.figure(figsize=(10, 5))
plt.title('Accumulated reward per battle')
plt.xlabel('Number of battles')
plt.ylabel('Accumalted Reward')
plt.plot(results)
plt.legend()
plt.show()
```

```python
#Load results
data = pd.read_json('VS Max Damage Player.json')
results = data['results']

#Plot results
plt.figure(figsize=(10, 5))
plt.title('Number of wins per 1000 battles')
plt.xlabel('Every 100 battles')
plt.ylabel('Battles')
plt.plot(results)
plt.legend()
plt.show()
```

```python
#Load results
data = pd.read_json('VS Max Damage Player.json')
results_max = data['results']

data = pd.read_json('VS Random Player.json')
results_random = data['results']

data = pd.read_json('Self Play.json')
results_self = data['results']

x = [*range(1, 11)]

#Plot results
plt.figure(figsize=(10, 5))
plt.xlabel('Every 100 out of 1000 games')
plt.ylabel('Number of wins')
plt.plot(x, results_max, label='Max Damage Player')
plt.plot(x, results_random, label='Random Player')
plt.plot(x, results_self, label='Self')
plt.legend()
plt.show()
```

# Final Results

```python
#Load all data
data_s0 = pd.read_json('Results/JSON/Reward Single Q-Learning: Type 0.json')
data_s1 = pd.read_json('Results/JSON/Reward Single Q-Learning: Type 1.json')
data_d0 = pd.read_json('Results/JSON/Reward Double Q-Learning: Type 0.json')
data_d1 = pd.read_json('Results/JSON/Reward Double Q-Learning: Type 1.json')

#Get results
results_s0 = data_s0['results']
results_s1 = data_s1['results']
results_d0 = data_d0['results']
results_d1 = data_d1['results']

#Plot results
plt.figure(figsize=(15, 10))
plt.title('Accumulated reward per battle')
plt.xlabel('Number of battles')
plt.ylabel('Accumalted Reward')
plt.plot(results_s0, label='Single Q-Learning: Type 0')
plt.plot(results_s1, label='Single Q-Learning: Type 1')
plt.plot(results_d0, label='Double Q-Learning: Type 0')
plt.plot(results_d1, label='Double Q-Learning: Type 1')
plt.legend()
plt.show()
```
